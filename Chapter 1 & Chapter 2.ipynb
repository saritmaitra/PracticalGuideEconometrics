## **Table 16. Transformation of time series data**

```python
import pandas as pd

gdp = pd.read_csv('/content/GDP.csv')

# DATE as the index and convert to datetime
gdp['DATE'] = pd.to_datetime(gdp['DATE'])
gdp.set_index('DATE', inplace=True)
gdp['Lag_1'] = gdp['GDP'].shift(1)

# natural logarithm of the GDP values.
gdp['log_gdp'] = np.log(gdp['GDP'])

# growth rate as the ratio of the current GDP to the lagged GDP.
gdp["growth_rate"] = gdp["GDP"] / gdp["Lag_1"]

"""subtracts 1 from this ratio to convert it to a growth rate in decimal form.
This growth rate represents the percentage change in GDP from one period to the next.
"""
gdp["growth_rate"] = gdp["GDP"] - 1

"""log approximation of the growth rate is used as it often provides a better linear
approximation for small changes, making it useful in regression models and other statistical analyses.
"""
gdp["log_approx"] = np.log(gdp["GDP"]) - np.log(gdp["Lag_1"])

gdp.dropna(inplace=True)
gdp.head(10)
```

## **Table 17. Reliability of natural logarithm**

```python
import numpy as np
import matplotlib.pyplot as plt

# series from 0 to 0.8 with a step of 0.01
g = np.arange(0, 0.8, 0.01)
log_g = np.log(1 + g)

# comparison of division and natural log approach
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))

# real growth rate vs. approximated growth rate
ax[0].plot(g, g, label="Division Approach")
ax[0].plot(g, log_g, label="Natural Log Approach")
ax[0].set_xlabel("Real Growth Rate", fontsize=16)
ax[0].set_ylabel("Approximated Growth Rate", fontsize=16)
ax[0].grid()
ax[0].legend(fontsize=12)

# difference between the two approaches
ax[1].plot(g, g - log_g, ls="--", lw=3, label="Difference Between Two Approaches")
ax[1].set_xlabel("Real Growth Rate", fontsize=16)
ax[1].set_ylabel("Difference (g - log_g)", fontsize=16)
ax[1].grid()
ax[1].legend(fontsize=16)

plt.show()
```

```python
import numpy as np
from scipy import stats

# synthetic data
np.random.seed(0)
group1 = np.random.normal(loc=0, scale=1, size=30)  # Group 1: mean=0, std=1
group2 = np.random.normal(loc=0, scale=2, size=30)  # Group 2: mean=0, std=2

# Perform Bartlett's test
statistic, p_value = stats.bartlett(group1, group2)

print("Bartlett's Test Statistic:", statistic)
print("p-value:", p_value)

# Interpretation
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The variances are significantly different.")
else:
    print("Fail to reject the null hypothesis: No significant difference in variances.")
```

### **Table 27. Chow test sample script**

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

np.random.seed(0)
X = np.random.rand(100, 1) * 10  # 100 random values as independent variable
Y = 2.5 * X + np.random.randn(100, 1) * 2  # Dependent variable with some noise

# Convert to DataFrame
data = pd.DataFrame({'X': X.flatten(), 'Y': Y.flatten()})

# Add a constant to the independent variable matrix for the intercept term
X_with_const = sm.add_constant(X)

# OLS model
model = sm.OLS(Y, X_with_const).fit()

# Chow Test (assuming a break at 50)
breakpoint = 50
y1 = Y[:breakpoint]
y2 = Y[breakpoint:]
X1 = X_with_const[:breakpoint]
X2 = X_with_const[breakpoint:]

# Fit models for both segments
model1 = sm.OLS(y1, X1).fit()
model2 = sm.OLS(y2, X2).fit()

# F-statistic for Chow Test
f_stat = ((model1.ssr + model2.ssr) / (model.ssr)) * (len(Y) - 2) / 2
print("Chow Test F-statistic:", f_stat)

from scipy import stats

# Degrees of freedom
df_num = 2  # Number of restrictions (2 models)
df_denom = len(Y) - 2  # Total observations minus number of parameters

# p-value
p_value = stats.f.sf(f_stat, df_num, df_denom)
print("Chow Test p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("The Chow Test indicates a significant structural break.")
else:
    print("The Chow Test does not indicate a significant structural break.")
```

### **Table 28. CUMSUM test script**

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import breaks_cusumolsresid

# sample data
np.random.seed(0)
X = np.random.rand(100, 1) * 10  # 100 random values as independent variable
Y = 2.5 * X + np.random.randn(100, 1) * 2  # Dependent variable with some noise

# Convert to DataFrame
data = pd.DataFrame({'X': X.flatten(), 'Y': Y.flatten()})

# Add a constant to the independent variable matrix for the intercept term
X_with_const = sm.add_constant(X)

# OLS model
model = sm.OLS(Y, X_with_const).fit()

# CUSUM Test using the residuals from the fitted model
cusum_result = breaks_cusumolsresid(model.resid)

# Print CUSUM results
print("CUSUM Test statistic:", cusum_result[0])
print("CUSUM Test p-value:", cusum_result[1])

# Interpretation
if cusum_result[1] < 0.05:
    print("CUSUM indicates parameter instability.")
else:
    print("CUSUM indicates parameter stability.")
```

### **Table 29. BDS test script**

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.stattools import bds

np.random.seed(0)
X = np.random.rand(100, 1) * 10  # 100 random values as independent variable
Y = 2.5 * X + np.random.randn(100, 1) * 2  # Dependent variable with some noise
# DataFrame
data = pd.DataFrame({'X': X.flatten(), 'Y': Y.flatten()})

# Add a constant to the independent variable matrix for the intercept term
X_with_const = sm.add_constant(X)

# OLS model
model = sm.OLS(Y, X_with_const).fit()

# BDS test on the residuals
bds_stat, p_value = bds(model.resid)

print("BDS Test statistic:", bds_stat)
print("BDS Test p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("BDS indicates nonlinearity in the residuals.")
else:
    print("BDS indicates linearity in the residuals.")
```

### **Table 30. Arch test**

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import het_arch

np.random.seed(0)
X = np.random.rand(100, 1) * 10
Y = 2.5 * X + np.random.randn(100, 1) * 2

# Convert to DataFrame
data = pd.DataFrame({'X': X.flatten(), 'Y': Y.flatten()})

# Add a constant to the independent variable matrix for the intercept term
X_with_const = sm.add_constant(X)

# OLS model
model = sm.OLS(Y, X_with_const).fit()

# Engle's ARCH test on the residuals
arch_test = het_arch(model.resid)

print("ARCH Test statistic:", arch_test[0])
print("ARCH Test p-value:", arch_test[1])

# Interpretation
if arch_test[1] < 0.05:
    print("ARCH test indicates the presence of heteroskedasticity.")
else:
    print("ARCH test indicates no significant heteroskedasticity.")
```
